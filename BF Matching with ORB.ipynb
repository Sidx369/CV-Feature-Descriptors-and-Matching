**Brute-Force Matching with ORB Descriptors**

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img1 = cv.imread('s1.png',cv.IMREAD_GRAYSCALE)        # queryImage
img2 = cv.imread('s2.png',cv.IMREAD_GRAYSCALE)        # trainImage
```

```python
# Initiate ORB detector
orb = cv.ORB_create()
```

```python
# find the keypoints and descriptors with ORB
kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)
```

```python
# create BFMatcher object
bf = cv.BFMatcher(cv.NORM_HAMMING, crossCheck=True)
```

```python
# Match descriptors.
matches = bf.match(des1,des2)
```

```python
# Sort them in the order of their distance.
matches = sorted(matches, key = lambda x:x.distance)
```

```python
# Draw first 10 matches.
img3 = cv.drawMatches(img1,kp1,img2,kp2,matches[:100],None,flags=cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(img3),plt.show()
```
