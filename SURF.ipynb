```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('s1.png',0)

# set Hessian Threshold to 400
surf = cv2.xfeatures2d.SURF_create(400)

# Find keypoints and descriptors directly
kp, des = surf.detectAndCompute(img,None)

len(kp)
```

```python
# Check present Hessian threshold
surf.getHessianThreshold()
```

```python
# Setting the threshold 3500 for representing in picture.
# In actual cases, it is better to have a value 300-500
surf.setHessianThreshold(3500)

# Compute keypoints and check its number of keypoints.
kp, des = surf.detectAndCompute(img,None)

len(kp)
```

```python
img2 = cv2.drawKeypoints(img,kp,None,(255,0,0),4)

plt.figure(figsize=(10,10))
plt.imshow(img2),plt.show()
```

```python
# Check upright flag, if it False, set it to True
surf.getUpright()
```

```python
surf.setUpright(True)
```

```python
# Recompute the feature points and draw
kp = surf.detect(img,None)
img2 = cv2.drawKeypoints(img,kp,None,(255,0,0),4)

plt.figure(figsize=(10,10))
plt.imshow(img2),plt.show()
```

```python
# Find size of descriptor
surf.descriptorSize()
```

```python
# flag, "extended" is False.
surf.getExtended()
```

```python
# Setting it to True to get 128-dim descriptors.
surf.setExtended(True)
kp, des = surf.detectAndCompute(img,None)
print(surf.descriptorSize())

print(des.shape)
```
